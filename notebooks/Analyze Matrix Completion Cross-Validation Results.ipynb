```python
def print_parameter_groups(df):
    parameter_groups = df.groupby(hyperparameter_names)
    parameter_auc = parameter_groups["auc"]
    parameter_f1 = parameter_groups["f1_score"]
    parameter_tau = parameter_groups["tau"]

    for (k, g) in parameter_groups:
        combined = 2 * (g["auc"] - 0.5) + g["f1_score"] + g["tau"]
        print("%s: min = %f, max = %f, median = %f, mean = %f" % (
                k,
                combined.min(),
                combined.max(),
                combined.median(),
                combined.mean()))
```

```python
print_parameter_groups(df_mean)
```

```python
print_parameter_groups(df_mice)
```

```python
allele_groups = df.groupby("allele")
for allele, g in allele_groups:
    for parameters, folds in g.groupby(hyperparameter_names):
        combined_score = (2 * (folds["auc"] - 0.5) + folds["tau"] + folds["f1_score"]).mean()
        print ("-- %s: %s %f" % (allele, parameters, combined_score))
    
```
